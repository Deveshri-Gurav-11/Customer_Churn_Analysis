```python
import pandas as pd

df = pd.read_excel("../dataset/Telco_customer_churn.xlsx")

df.head()
```

```python
df.columns
```

```python
df.shape
df.info()
df['Churn Value'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Churn Label', data=df)

plt.title("Customer Churn Distribution")

plt.show()
```

Customer Churn Distribution Insight
- The dataset contains both retained and churned customers.
- A significant number of customers have churned, indicating potential retention issues.
- Understanding the factors behind churn can help businesses improve customer retention strategies.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

sns.countplot(x='Contract', hue='Churn Label', data=df)

plt.title("Customer Churn by Contract Type")

plt.show()
```

Churn by Contract Type Insights
- Customers with month-to-month contracts churn more frequently.
- Customers with 1-year or 2-year contracts are more likely to stay.

```python
plt.figure(figsize=(8,5))

sns.boxplot(x='Churn Label', y='Tenure Months', data=df)

plt.title("Tenure vs Customer Churn")

plt.show()
```

Customers with lower tenure are more prone to churn, while long-term customers demonstrate higher retention and loyalty.

```python
plt.figure(figsize=(8,5))

sns.boxplot(x='Churn Label', y='Monthly Charges', data=df)

plt.title("Monthly Charges vs Customer Churn")

plt.show()
```

Monthly Charges VS Customer Churn Insight
- Customers with higher monthly charges show higher churn rates.
- Pricing strategy may influence customer retention.

```python
plt.figure(figsize=(8,5))

sns.countplot(x='Internet Service', hue='Churn Label', data=df)

plt.title("Churn by Internet Service Type")

plt.show()
```

Churn By Internet Service Type Insight
- Some service types may have higher churn than others.
- This may indicate service quality or pricing issues.

Overall Insights
- Customers with month-to-month contracts churn the most.
- Short-tenure customers are more likely to leave.
- Higher monthly charges increase churn probability.
- Long-term contracts help improve customer retention.
